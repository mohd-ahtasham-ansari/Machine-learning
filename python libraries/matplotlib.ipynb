# Matplotlib 

useful for making graphs

```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# importing numpy to get data for our plots
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
# sine wave
plt.plot(x,y)
plt.show
```

```python
# cosine wave
plt.plot(x,z)
plt.show
```

```python
# adding title x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('Angles')
plt.ylabel('Sine value')
plt.title('Sine wave')
plt.show
```

lets plot a parabola

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show
```

```python
# lets plot in another way
plt.plot(x,y,'r+')
plt.show
```

```python
plt.plot(x,y,'g.')
```

```python
x=np.linspace(-5,5,70)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'g--')
plt.show
```

PLOT bar graph

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['english', 'hindi', 'punjabi','urdu', 'marathi','tamil']
people=[100,110,34,35,40,38]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('number of people')
plt.show
```

```python
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,0.5])
subjects=['maths','physics ','chemistry','biology','english','additional_sub']
marks=[92,90,45,98,99,100]
ax.bar(subjects,marks)
plt.xlabel('SUBJECT')
plt.ylabel('Marks Obtained')
plt.show
```

PIE chart

```python
fig3=plt.figure()
ax=fig3.add_axes([0,0,1,1])
languages=['english', 'hindi', 'punjabi','urdu', 'marathi','tamil']
people=[100,110,34,35,40,38]
ax.pie(people,labels=languages,autopct='%1.1f%%')

plt.show
```

Scatter Plot

```python
# lets plot a scattter plot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig5=plt.figure()
ax=fig5.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show
```

3D scatter Plot

```python
fig6=plt.figure()
ax=plt.axes(projection='3d')
z= 20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show
```
